## Brief R square analysis

This brief analysis shows how the harmonic parameter influences the `R_square` score.

This plot used the data sampled from this [GEE script](https://code.earthengine.google.com/749249a98a9f4a584d2491ef767ed72c). The output from this script are exported to the Drive, cleaned up and pre-process (to make them ready to use from `pandas` module). The cleaned up data is available in the  `Interpolation\sample_data.xlsx` file. 

Several points are important to notice:
- The data has been sampled from 3 different locations (North, center and South of the UK, see Map annexes) at 3 different periods (January, June and September 2018). There are 9 images.
- The interpolation is performed over the whole 2018 date range.
- The number of images used for the interpolation is different between the North image and the two others. The two images from the center and the south have very few data between January and June (less than 10). This is because I have exported all the image from the GEE storage (no place left). However, I kept all the image in the north. That means, this is the most representative between the 3 locations.

- The number of points per image sampled is different according the number of harmonic used. I tried to fix it with the `seed` paramters in the `sample` methods but the problem persists. I do not understand why.

- The *R-square* score is computed using the [`r2_score`](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html). This method can return negative values if the model is very bad as explained in the documentation. 



```python
# Import modules
import pandas as pd                      # Handle dataset
from sklearn.metrics import r2_score     # Compute r square score
import re                                # Regex : extract number from string
import xlrd                              # Handle excel file: extract sheet names
import matplotlib.pyplot as plt          # Plots
import seaborn as sns                    # Plots styles

# Init plot styles
sns.set()
```

```python
# Filename dataset
filename = r'Interpolation/sample_data.xlsx'

# Get sheet names (images)
xls = xlrd.open_workbook(filename, on_demand=True)
sheet_names = xls.sheet_names() 

out = []
# For each sheet (images)
for sheet in sheet_names:
    # Read excel sheet content
    df = pd.read_excel(filename, sheet_name=sheet)
    cols = df.columns
#     print(sheet)
    
    # Compute R square score for each number of harmonic
    r_squares = []
    for true, pred in zip(cols[::2], cols[1::2]):
        # Ignore empty columns (raising errors)
        if df[true].dropna().shape[0] > 0:
            sub_dict = {"n_values": df[true].dropna().shape[0],
                        "R_square": r2_score(df[true].dropna(), df[pred].dropna()),
                        "Nb_harmo": int(re.findall(r'\d+', true)[0])}
            r_squares.append(sub_dict)
    out.append({sheet: r_squares})
```

```python
fig, axs = plt.subplots(1, 3, figsize=(13,10))
index = 0
color = 'tab:red'
for o in out:
    if "north" in list(o.keys())[0]:
        title = list(o.keys())[0]
        values = list(o.values())[0]
        results = pd.DataFrame().from_dict(values).sort_values(by='Nb_harmo')
        
        ax_twin = axs[index].twinx()
        l1 = axs[index].plot(results["Nb_harmo"], results["R_square"], color=color)
        l2 = ax_twin.plot(results["Nb_harmo"], results["n_values"])
        
        leg = l1 + l2
        labs = [l.get_label() for l in leg]
        axs[index].legend(leg, labs, loc='lower center')
        
        axs[index].set_ylabel(r"$R^{2}$")
        ax_twin.set_ylabel(r"Number of points used", rotation=270)
        axs[index].set_title(title)
        
        index += 1
        
fig.suptitle("R square VS the number of harmonics", size=15)
fig.tight_layout()
fig.subplots_adjust(top=0.92)
```

### R square score on the others images (Center and South). 

The following plots shows that a few number of image used leads to strange results (unpredictable ?).

```python
fig, axs = plt.subplots(2, 3, figsize=(13,10))
index = 0
for o in out:
    if "north" not in list(o.keys())[0]:
        title = list(o.keys())[0]
        values = list(o.values())[0]
        results = pd.DataFrame().from_dict(values).sort_values(by='Nb_harmo')
        
        axs[index//3, index%3].set_title(title)

        
        ax_twin = axs[index//3, index%3].twinx()
        l1 = axs[index//3, index%3].plot(results["Nb_harmo"], results["R_square"], color=color)
        l2 = ax_twin.plot(results["Nb_harmo"], results["n_values"])
        
        leg = l1 + l2
        labs = [l.get_label() for l in leg]
        axs[index//3, index%3].legend(leg, labs, loc='lower center')
        
        axs[index//3, index%3].set_ylabel(r"$R^{2}$")
        ax_twin.set_ylabel(r"Number of points used", rotation=270)
        
        index += 1
        
fig.suptitle("R square VS number of harmonics", size=15)
fig.tight_layout()
fig.subplots_adjust(top=0.92)
```

## Discussion

- The range of the analysis directly impacts the best values of the number of harmonics. The "North" images (with the most data) globally shows the bigger the number of harmonics, the better the R square. However, this solution implies having a lot of images. Doing this process over all the UK isn't possible due to the GEE restrictions.

- However, if some images are missing in the interval, increasing the number of harmonics seems to degrade the R square score. The following graph shows the interpolations for the points 2. It also shows the gap between January and July. this missing data might explains the divergent points.
